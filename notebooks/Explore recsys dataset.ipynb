```python
metadata_features_genre = pd.read_csv("../data/recsys/Metadata/GenreFeatures.csv", index_col=0)
```

```python
metadata_features_tag = pd.read_csv("../data/recsys/Metadata/TagFeatures.csv", index_col=0)
```

```python
metadata_features_tag.index
```

```python
metadata_features_genre = metadata_features_genre.loc[metadata_features_tag.index]
```

```python
data = [metadata_features_tag, metadata_features_genre]
data = pd.concat(data, axis=1)
```

```python
groups = metadata_features_tag.index
```

```python
latent_dim = 1024   
config = {
    "encoder": [
        {
            "kwargs": {
                "activation": "relu",
                "units": 2048
            },
            "name": "hidden1",
            "type": "Dense"
        },
        {
            "name": "batchnorm1",
            "type": "BatchNormalization"
        },
        {
            "kwargs": {
                "rate": 0
            },
            "name": "dropout",
            "type": "Dropout"
        },
        {
            "kwargs": {
                "activation": "relu",
            },
            "name": "latent",
            "regularizer": {
                "type": "l1",
                "value": 0
            },
            "type": "Dense"
        }
    ]
}
```

```python
latent_shape = (latent_dim,)
input_shape = (data.shape[1],)
```

```python
ae = Autoencoder(config["encoder"],
                 None,
                 input_shape=input_shape,
                 latent_shape=latent_shape,
                 loss="mean_absolute_error",
                 optimizer_params=None)
```

```python
scores = ae.run_training(data, groups, random_state=100, epochs=1000, standardize=True, callbacks = [kc.EarlyStopping(monitor="val_loss", min_delta=0.000001, patience=10)])
```
